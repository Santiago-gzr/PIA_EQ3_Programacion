```python
import datetime as dt  #Importamos la libreria de fechas

fecha_hoy=dt.date.today()  #Asignamos la fecha de hoy

def es_fecha_valida(date_string):  #Asignamos la funcion que valida la fecha sea correcta
    try:
        dt.datetime.strptime(date_string, "%d-%m-%Y")
        return True
    except ValueError:
        return False


oficios=[] #<----oficios requeridos aqui

oficios.append("Carpintero")  #aqui insertamos los oficios requeridos
oficios.append("Cerrajero")
oficios.append("Mecánico")
oficios.append("Pescador")
oficios.append("Albañil")
oficios.append("Fontanero")
oficios.append("Carpintero")
oficios.append("Soldador")
oficios.append("Pintor")
oficios.append("Sastre")


#Iniciamos el ciclo de preguntas
while True:
  parental=input("¿Su familia es monoparental o biparental? (Mono/Bi) o deje en blanco para salir: ").upper()  #Pregutamos si la familia en mono o bi
  puntos=0        #Empezamos con los contadores en 0
  contador_h=0    #Y los metemos en el while porque se reiniciaran cada ves que termine de registrar una familia
  if parental=="":
   break
  if parental=="BI":    #Dependiendo de la respuesta prenguntamos 2 veces los datos de los padres
    pts=0      #Empezamos con los contadores en 0
    pts2=0     #Y los metemos en el while porque se reiniciaran cada ves que termine de registrar una familia
    fecha_nacimiento_primer_progenitor = None
    grado_academico_primer_progenitor = None
    oficios_primer_progenitor = None
    while True:
     sexo_primer_progenitor=input("¿Cual es el sexo del progenitor? (M/F/N): ").upper()  #Primero preguntamos el sexo
     if sexo_primer_progenitor in ["F", "M", "N"]:
      while True:
       fecha_nacimiento_primer_progenitor=input("¿Cual es la fecha de nacimiento del progenitor? (dd-mm-aaaa): ")  #Y ahora la fecha de nacimiento
       if (
       len(fecha_nacimiento_primer_progenitor) == 10
       and fecha_nacimiento_primer_progenitor[2] == fecha_nacimiento_primer_progenitor[5] == "-"
       and fecha_nacimiento_primer_progenitor[0:2].isdigit()
       and fecha_nacimiento_primer_progenitor[3:5].isdigit()                                       #Aqui validamos que la fecha sea correcta
       and fecha_nacimiento_primer_progenitor[6:].isdigit()
       and es_fecha_valida(fecha_nacimiento_primer_progenitor)
       ):
         fecha_nacimiento = dt.datetime.strptime(fecha_nacimiento_primer_progenitor, "%d-%m-%Y").date()  #Pasamos la fecha a dato de fecha

         fecha_actual = dt.date.today()
         limite_edad = 150

         if fecha_nacimiento <= fecha_actual and fecha_actual.year - fecha_nacimiento.year <= limite_edad:  #Aqui validamos que la edad sea coherente
            if sexo_primer_progenitor=="F":                        #Y dependiendo de si es Mujer calculamos la edad
              edad=(fecha_hoy.year)-(fecha_nacimiento.year)
              if edad>20 and edad<35:
                pts2=(15-(edad-20))       #Si esta en la edad fertil de hace la resta de los años que le quedan y se le suman los puntos correspondientes
                puntos+=pts2
              if edad<=20:
                puntos+=15              #Y si es menos a 20 se da por hecho que le queda toda la edad fertil y se le suma 15 puntos
              break
            if sexo_primer_progenitor=="M":
              break

         else:
           print("Error, fecha de nacimiento no válida vuelva a intentar.")
                                                                 #En todo esto validamos si algo sale mal nos de un mensaje de error
       else:
        print("Error, valor no valido. Por favor, ingrese la fecha en el formato 'dd-mm-aaaa'.")
        continue
     else:
      print("Error, valor no valido vuelva a intentar.")
      continue
     break

    while True:        #Preguntamos el grado academico
     grado_academico_primer_progenitor=input("¿Cual es el grado academico del progenitor? (Postgraduado=P, Universitario=U, Tecnico=T, Ninguno=N: ").upper()
     if grado_academico_primer_progenitor=="P":        #Y validamos que ingrese los datos de manera correcta
      pts+=5                      #Si es P se le suma 5 puntos
      break
     if grado_academico_primer_progenitor=="U":
      pts+=3                      #Si es U se le suma 3 puntos
      break
     if grado_academico_primer_progenitor=="T":
      break
     if grado_academico_primer_progenitor=="N":
      break
     else:
      print("Error en la captura vuelva a intentar")   #Si algo sale mal sale el mensaje de Error
      continue
    print(oficios)      #Primero los imprimimos para que se vean cuales son los requeridos
    while True:          #Preguntamos si tiene un oficio requerido
     oficios_primer_progenitor=input("¿Tiene alguno de los 10 oficios requeridos en el pais?  (S/N): ").upper()
     if oficios_primer_progenitor=="S":
      pts+=8
      break
     if oficios_primer_progenitor=="N":
      break
     else:
      print("Error en la captura, intente de nuevo")
      continue
    puntos+=pts       #Y aqui simplemente vuelve a preguntar los mismos datos al segundo progenitor
    pts=0
    pts2=0             #Y se hace exactamente los mismo
    fecha_nacimiento_primer_progenitor = None
    grado_academico_primer_progenitor = None
    oficios_primer_progenitor = None
    while True:
     sexo_primer_progenitor=input("¿Cual es el sexo del progenitor? (M/F/N): ").upper()
     if sexo_primer_progenitor in ["F", "M", "N"]:
      while True:
       fecha_nacimiento_primer_progenitor=input("¿Cual es la fecha de nacimiento del progenitor? (dd-mm-aaaa): ")
       if (
       len(fecha_nacimiento_primer_progenitor) == 10
       and fecha_nacimiento_primer_progenitor[2] == fecha_nacimiento_primer_progenitor[5] == "-"
       and fecha_nacimiento_primer_progenitor[0:2].isdigit()
       and fecha_nacimiento_primer_progenitor[3:5].isdigit()
       and fecha_nacimiento_primer_progenitor[6:].isdigit()
       and es_fecha_valida(fecha_nacimiento_primer_progenitor)
       ):
         fecha_nacimiento = dt.datetime.strptime(fecha_nacimiento_primer_progenitor, "%d-%m-%Y").date()

         fecha_actual = dt.date.today()
         limite_edad = 150

         if fecha_nacimiento <= fecha_actual and fecha_actual.year - fecha_nacimiento.year <= limite_edad:
            if sexo_primer_progenitor=="F":
              edad=(fecha_hoy.year)-(fecha_nacimiento.year)
              if edad>20 and edad<35:
                pts2=(15-(edad-20))
                puntos+=pts2
              if edad<=20:
                puntos+=15
              break

         else:
           print("Error, fecha de nacimiento no válida vuelva a intentar.")

       else:
        print("Error, valor no valido. Por favor, ingrese la fecha en el formato 'dd-mm-aaaa'.")
        continue
     else:
      print("Error, valor no valido vuelva a intentar.")
      continue
     break

    while True:
     grado_academico_primer_progenitor=input("¿Cual es el grado academico del progenitor? (Postgraduado=P, Universitario=U, Tecnico=T, Ninguno=N: ").upper()
     if grado_academico_primer_progenitor=="P":
      pts+=5
      break
     if grado_academico_primer_progenitor=="U":
      pts+=3
      break
     if grado_academico_primer_progenitor=="T":
      break
     if grado_academico_primer_progenitor=="N":
      break
     else:
      print("Error en la captura vuelva a intentar")
      continue
    print(oficios)
    while True:
     oficios_primer_progenitor=input("¿Tiene alguno de los 10 oficios requeridos en el pais?  (S/N): ").upper()
     if oficios_primer_progenitor=="S":
      pts+=8
      break
     if oficios_primer_progenitor=="N":
      break
     else:
      print("Error en la captura, intente de nuevo")
      continue
    puntos+=pts
  if parental=="MONO":     #Dependiendo de la respuesta prenguntamos 1 vez los datos del padre
    pts=0
    pts2=0                #Se hace el mismo procedimientos pero solo 1 vez
    fecha_nacimiento_primer_progenitor = None
    grado_academico_primer_progenitor = None
    oficios_primer_progenitor = None
    while True:
     sexo_primer_progenitor=input("¿Cual es el sexo del progenitor? (M/F/N): ").upper()
     if sexo_primer_progenitor in ["F", "M", "N"]:
      while True:
       fecha_nacimiento_primer_progenitor=input("¿Cual es la fecha de nacimiento del progenitor? (dd-mm-aaaa): ")
       if (
       len(fecha_nacimiento_primer_progenitor) == 10
       and fecha_nacimiento_primer_progenitor[2] == fecha_nacimiento_primer_progenitor[5] == "-"
       and fecha_nacimiento_primer_progenitor[0:2].isdigit()
       and fecha_nacimiento_primer_progenitor[3:5].isdigit()
       and fecha_nacimiento_primer_progenitor[6:].isdigit()
       and es_fecha_valida(fecha_nacimiento_primer_progenitor)
       ):
         fecha_nacimiento = dt.datetime.strptime(fecha_nacimiento_primer_progenitor, "%d-%m-%Y").date()

         fecha_actual = dt.date.today()
         limite_edad = 150

         if fecha_nacimiento <= fecha_actual and fecha_actual.year - fecha_nacimiento.year <= limite_edad:
            if sexo_primer_progenitor=="F":
              edad=(fecha_hoy.year)-(fecha_nacimiento.year)
              if edad>20 and edad<35:
                pts2=(15-(edad-20))
                puntos+=pts2
              if edad<=20:
                puntos+=15
              break
            if sexo_primer_progenitor=="M":
              break

         else:
           print("Error, fecha de nacimiento no válida vuelva a intentar.")

       else:
        print("Error, valor no valido. Por favor, ingrese la fecha en el formato 'dd-mm-aaaa'.")
        continue
     else:
      print("Error, valor no valido vuelva a intentar.")
      continue
     break

    while True:
     grado_academico_primer_progenitor=input("¿Cual es el grado academico del progenitor? (Postgraduado=P, Universitario=U, Tecnico=T, Ninguno=N: ").upper()
     if grado_academico_primer_progenitor=="P":
      pts+=5
      break
     if grado_academico_primer_progenitor=="U":
      pts+=3
      break
     if grado_academico_primer_progenitor=="T":
      break
     if grado_academico_primer_progenitor=="N":
      break
     else:
      print("Error en la captura vuelva a intentar")
      continue
    print(oficios)
    while True:
     oficios_primer_progenitor=input("¿Tiene alguno de los 10 oficios requeridos en el pais?  (S/N): ").upper()
     if oficios_primer_progenitor=="S":
      pts+=8
      break
     if oficios_primer_progenitor=="N":
      break
     else:
      print("Error en la captura, intente de nuevo")
      continue
    puntos+=pts
  if parental=="BI" or parental=="MONO":
    while True:                           #Aqui entra el ciclo de preguntas de los hijos
      hijos=input("Hay hijos en su familia (S/N): ").upper()
      if hijos=="S":          #Si responde que si sucede esto
        if parental=="BI":
          puntos+=12           #Si la respuesta fue bi suma 12 ya que a cada progenitor suma 6
        if parental=="MONO":
          puntos+=6             #Si la respuesta fue mono suma 6 ya que a cada progenitor suma 6
        while True:
          try:
            cant_hijos=input("Cuantos hijos hay en su familia: ")    #Preguntamos la cantidad
            cantidad_hijos=int(cant_hijos)
            cantidad_hijos>0 and cantidad_hijos<10         #Y validamos que sea numerico el dato
            break
          except:
            print("Error al capturar el numero vuelva a intentar")  #Si sale mal aparece un mensaje de Error
        puntos+=(cantidad_hijos*8)    #Cada hijo suma 8 asi que dependiendo de la cantidad de hijos lo multiplica por 8 y suma los puntos correspondientes
        while True:              #Empieza un ciclo
          if (contador_h)==(cantidad_hijos):   #Y dependiendo de la cantidad de hijos es la cantidad de vueltas que va dar el ciclo
            break
          contador_h+=1
          while True:
            sexo_hijo=input("Ingrese la clave de sexo del hijo (F, M, N): ").upper()  #Preguntamos el sexo del hijo
            if sexo_hijo=="F":
             break
            if sexo_hijo=="M":             #validamos que sea correcto el dato
             break
            if sexo_hijo=="N":
             break
            else:
             print("No se reconoce la accion")    #Si sale mal da un mensaje de Error
             continue
          while True:
            try:
             fecha_hijo=input("Ingrese la fecha de nacimiento del hijo (dd-mm-aaaa): ")  #Preguntamos la fecha de ancimiento del hijo
             diah=int(fecha_hijo[0:2])
             mesh=int(fecha_hijo[3:5])
             añoh=int(fecha_hijo[6:])
             fecha_h=dt.date(añoh,mesh,diah)         #Validamos que la fecha sea correcta
             break
            except:
             print("La fecha es incorrecta, intentelo de nuevo")  #Si sale mal da un mensaje de Error
             continue
          if sexo_hijo=="F":        #Si el hijo es mujer
             edad_h=(fecha_hoy.year)-(fecha_h.year)   #Calculamos la edad
             if edad_h>20 and edad_h<35:
              pts_h=(15-(edad_h-20))    #Si la hija tiene mas de 20 y menos de 35 calculamos la edad fertil que le quda
              puntos+=pts_h
             if edad_h<=20:       #Si tiene menos o igual a 20 se le suman 15 puntos ya que se da por hecho que le que toda la edad fertil
              puntos+=15
        print(f"La cantidad de puntos para su familia es {puntos}")  #Se imprimen los puntos que acumulo la Familia
        break         #Y a qui termina el ciclo de preguntas
      if hijos=="N":  #Si no tiene hijos
        print(f"La cantidad de puntos para su familia es {puntos}")  #Se imprimen los puntos que acumulo la Familia
        break         #Y a qui termina el ciclo de preguntas
      else:           #Si algo sale mal con la captura de hijos de Si o No
        print("Error no se reconoce la accion vuelva a intentar")  #Sale el mensaje de error
        continue
  else:             #Si algo sale mal con la captura de Bi o Mono
    print("Error en la captura, vuelva a intentar")   #Sale el mensaje de error
    continue    #Y vuelve a preguntar
```
